````python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16, 8)

df = pd.read_csv('lønn.csv', sep=";")

df.head()
````

````python
df.plot(x="År", y=["Sykepleiere", "Lektorer vgs", "Allmennpraktiserende leger"])
plt.show()
````

````python
from scipy import stats

res = stats.linregress(df["År"], df["Lektorer vgs"]-df["Sykepleiere"])
plt.plot(df["År"], res.intercept + res.slope*df["År"], 'r', label='fitted line')
plt.plot(df["År"], df["Lektorer vgs"]-df["Sykepleiere"])
plt.show()
print(res.slope)
````
